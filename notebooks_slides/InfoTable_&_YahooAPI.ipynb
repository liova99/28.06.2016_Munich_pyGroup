# Yahoo API

```python
# http://chartapi.finance.yahoo.com/instrument/1.0/tsla/chartdata;type=quote;range=1m/csv
# Pandas power
import pandas.io.data as web
import datetime
```

```python
# Define start date, format Y, M, D 
start = datetime.datetime(2010, 3, 1)

# pandas data_reader df = web.get_data_yahoo( 'TSLA', start, end, interval='w' )
df = web.get_data_yahoo('AAPL', start) #TSLA for Tesla
df.tail()
```

```python
from bokeh.plotting import figure, output_notebook, show
output_notebook()
```

```python
TOOLS = 'pan,wheel_zoom,box_zoom,crosshair,resize,reset'


f = figure(height=370, width=800, x_axis_type="datetime", tools=TOOLS)#, responsive = True)

# plot type (x_axis, y_axis, + many configurations)
f.line(df.index, df.Open, color='blue')
#f.circle(df.index, df.Open, color='navy', size=0.5, alpha=0.8)

show(f)
```
